## 3. Format Data

```python
feature_current=list(df_current.columns.values)
feature_2016=list(df_2016.columns.values)
print("feature_current are:", feature_current)
print("feature_2016 are:", feature_2016)
```

### 3.1 Convert to Lower Case

```python
def lowercase(x):
    for i in range (len(x)):
        x[i] = str.lower(str(x[i]))
```

```python
lowercase(feature_current)
lowercase(feature_2016)
print("feature_current are:", feature_current)
print("feature_2016 are:", feature_2016)
```

### 3.2 Delete underscores

```python
def del_underscore(x):
    for i in range(len(x)):
        x[i]=str(x[i]).replace('_',"")
```

```python
del_underscore(feature_2016)
print("feature_current are:", feature_current)
print("feature_2016 are:", feature_2016)
```

### 3.3 Get common features

```python
feature_common = (set(feature_current))& set(feature_2016)
print (len(feature_common),"\nCommon features are:", sorted(feature_common))
```

### 3.4 Understand unique features

```python
unique_2016=set(feature_2016)-set(feature_common)
unique_current=set(feature_current)-set(feature_common)
print (len(unique_current), sorted(unique_current))
print (len(unique_2016), sorted(unique_2016))
```

### Find common features with different names

```python
# type(feature_common), set
len(feature_common) #100
```

```python
# current: addrzip, fundedamount, loanamount, numacctsever120ppd, reviewstatus,     , reviewstatusd
# 2016:    zipcode, fundedamnt,   loanamnt  , numacctsever120pd,  verificationstatus, verificationstatusjoint
```

```python
feature_common=feature_common.union({'zipcode', 
 'fundedamnt',
 'loanamnt',
 'verificationstatus',
 'verificationstatusjoint',
 'numacctsever120pd'})
len(feature_common)
```

### Replace df_2016 with new common features

```python
for col in df_2016.columns.values:
    df_2016.rename(columns={str(col):str(col).replace('_','')}, inplace=True)
df_2016.columns
```

```python
#df[list(common features)+[new col]]
df_2016_new=df_2016[list(feature_common)+['issued', 'loanstatus']]
df_2016_new.shape
```

### 3.5 Create Feature Dictionary

```python
feature_dic={}
feature_dic['feature_common']=feature_common
feature_dic['unique_2016']=unique_2016
feature_dic['unique_current']=unique_current
feature_dic
```
